# TIC TAC TOE

```python
from tkinter import *

placement = {"1":(0,0), '2':(0,1), '3':(0,2), '4':(1,0), '5':(1,1), '6':(1,2), '7':(2,0), '8':(2,1), '9':(2,2)}
board = [["-" for i in range(3)] for j in range(3)]
gameover = False
i = 0

def is_filled_2d_array(arr):
    #checks if the board is filled
    return all(cell != "-" for row in arr for cell in row)

def print_board(board, i):
    for row in board:
        i += 1
        if i!=3:
            print(" | ".join(map(str,row)))
            print("---------")
        else:
            print(" | ".join(map(str,row)))


def check_game(board):
    #check rows:
    for row in range(3):
        if board[row][0]==board[row][1] and board[row][2]==board[row][1] and board[row][0] != "-":
            print("GAME OVER! " + board[row][0] + " WON!")
            return True
            
    #check columns:
    for col in range(3):
        if board[0][col]==board[1][col] and board[1][col]==board[2][col] and board[0][col] != "-":
            print("GAME OVER! " + board[0][col] + " WON!")
            return True
            
    #check diagonals:
    if board[0][0]==board[1][1] and board[1][1]==board[2][2] and board[1][1]!='-':
        print("GAME OVER! " + board[1][1] + " WON!")
        return True
        
    if board[0][2]==board[1][1] and board[1][1]==board[2][0] and board[1][1]!="-":
        print("GAME OVER! " + board[1][1] + " WON!")
        return True
        
    #check full:
    is_filled_2d_array(board)

    return False
    


gameover = False
i = 0

while gameover == False:
    #PLAYER X MOVES
    X_move = input("playerX choose a position on the board (1-9): ")
    if X_move in placement:
        x,y = placement[X_move]
        if board[x][y] == "-":
            board[x][y] = "X"
        else:
            while board[x][y] != "-":   
                X_move = input("spot taken, please choose another one, playerX: ")
                x,y = placement[X_move]
                if board[x][y] == "-":
                    board[x][y] = "X"
    else:
        X_move = input("invalid input. Select a number from 1-9")
    
    #print the board
    print_board(board, i)
    i = 0
    #check if the game is over
    gameover = check_game(board)
    if gameover == True:
        break

    #print empty line
    print("\n")
    
    #PLAYER O MOVES
    O_move = input("playerO choose a position on the board (1-9): ")
    if O_move in placement:
        x,y = placement[O_move]
        if board[x][y] == "-":
            board[x][y] = "O"
        else:
            while board[x][y] != "-":   
                O_move = input("spot taken, please choose another one, playerO: ")
                x,y = placement[O_move]
                if board[x][y] == "-":
                    board[x][y] = "O"
    else:
        O_move = input("invalid input. Select a number from 1-9")

    #print the board
    print_board(board, i)
    i = 0

    #check if the game is over
    gameover = check_game(board)
    if gameover == True:
        break

    #print empty line
    print("\n")


    
    


```
